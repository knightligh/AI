```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
file ='/1.01. Simple linear regression (1).csv'
data = pd.read_csv(file)
data.head(21)
x1=np.array(data['SAT'])
y=np.array(data['GPA'])

#np.reshape(1, -1)
plt.scatter (x1,y)
plt.xlabel('SAT',fontsize =20 )
plt.ylabel('GPA',fontsize=20)
plt.show()
from sklearn import linear_model
regression_model = linear_model.LinearRegression()
regression_model.fit(x1,y)

print(regression_model.fit_intercept)
```

```python
#bai tap 2 
#from pandas as pd 
url='/mtcars.csv'
data=pd.read_csv(url)
data.hist()
plt.show()
correlation = data.corr()
print(correlation)
X = data.drop(data.columns[[0,2]],axis=1)
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black')

```

```python
#bai tap 3 hồi quy đa biến 
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['wt']),y=data['mpg'])
print(model.intercept_)
print(model.coef_)
model.score(pd.DataFrame(data['wt']),data['mpg'])# phan tram do tin cay
y_pred = model.predict(X=pd.DataFrame(data['wt']))
versidual = data['mpg']-y_pred
versidual.describe()

data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='red')
plt.plot(data['wt'],y_pred,color='blue')
```

```python
#bai tap 4 
#DU LIEU HE BAC 2 
y=data['mpg']
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'],data['wt']**2]).T
model.fit(X,y=data['mpg'])
print('Intercept cua no la hinh bac 2:')
print(model.intercept_)
print('coffient cua hinh bac 2 :')
print(model.coef_)
print('do chinh xac cua mo hinh :')
print(model.score(X,y))
```

```python
#bai tap so 5 
#dữ liệu bậc 2
range = np.arange (1.5,5.5,0.1)
predictor = pd.DataFrame([range,range**2]).T
y_pred = model.predict (X=predictor)
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='red',xlim=(0,7))
plt.plot(range,y_pred,color='blue')
```

```python
#II/phân tích dữu liệu của start-up
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd 
import seaborn as sns

data=pd.read_csv ('/50_Startups.csv')
data.head 
#hiển tị theo cặhp dữ liệu 
sns.pairplot(data)

corr=data.corr()
sns.heatmap(corr,annot=True)

X =data.drop('Profit',axis=True)#xoá cột dữ liệu profit
X =data.drop('State',axis=True)#xoá cột dữ  liệu state 
y=data['Profit']
X.head()

from sklearn.preprocessing import StandardScaler#tỉ lệ hoá số liệu (normal line số liệu)
SC=StandardScaler()
X=SC.fit_transform(X)

from sklearn.model_selection import train_test_split
X_train ,X_test,y_train ,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
X_train.shape
X_test.shape
y_train.shape
y_test.shape
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
y_pred = regressor.predict(X_test).round(1)
caculation = pd.DataFrame(np.c_[y_test,y_pred],columns=['Original Profit','Predicted Profit'])
print('do chinh xac cua mo hinh :')
print(regressor.score(X_train,y_train))
```
