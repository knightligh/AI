```python
#nhận diện con vật từ dữ liệu Cifar 10

from keras.datasets import cifar10
import matplotlib.pyplot as plt
(x_train,y_train),(x_test,y_test)=cifar10.load_data()
x_train.shape
y_train.shape
import matplotlib.pyplot as plt
for i in range(10):
  plt.subplot(3,5,i+1)
  plt.tight_layout()
  plt.imshow(x_train[i],cmap='gray',interpolation='None')
  plt.title('Number: {}'.format(y_train[i]))
  plt.xticks([])
  plt.yticks([])

x_train=x_train.reshape((50000,32*32*3))#chinh lai kich thuoc au du lieu dau vao
x_train_1= x_train.astype('float32')/255 # chuyen qua float de co the chia duoc

x_test=x_test.reshape((10000,32*32*3))
x_test_1=x_test.astype('float32')/255

from keras.utils import to_categorical #chuyen thanh one_hot_encoding
y_train_1= to_categorical(y_train,10)# chyen thanh 10 bit one hot encoding
y_test_1= to_categorical(y_test,10)
print(y_test)

#Tao ra mang Neural
from keras.models import Sequential
from keras.layers import Dense
from keras import models  

model=models.Sequential()
model.add(Dense(512,input_shape=(32*32*3,),activation="relu"))
model.add(Dense(256,activation="relu"))
model.add(Dense(128,activation="relu"))

model.add(Dense(10,activation="linear",name="Softmax"))
model.summary

model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
model.fit(x_train_1,y_train_1,epochs=10,batch_size=128)#qua trinh train, moi lan hoc 128 voi 10 lan hoc lai
test_lost,test_acc=model.evaluate(x_test,y_test)
print('test_accuracy',test_acc)
print('test_lost',test_lost)

```
